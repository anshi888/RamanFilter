```python
import pandas as pd
filename = '/content/Raman1 - Punto 1 (1).csv'
data=pd.read_csv(filename,delimiter=',',decimal=",")
print(data)
```

```python
from numpy import *

x= data.X
y= data.Y
#pasar datos
x1= []
y1=[]
#print(x)
for c in range(len(x)):
    x1.append(float(x[c]))
    y1.append(float(y[c]))
#print(type(x1))
#print(x1)
```

```python
from numpy import *
import matplotlib.pyplot as plt
# Y los graficamos
plt.figure(figsize=(26,10))
plt.title('Datos con ruido')
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim(200,3500)
plt.plot(x1, y1)
```

```python
from numpy import *
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter as savgol

yf1 =  savgol(data.Y,101,0)

for u in range(len(y1)):
  data.Y[u] = yf1[u]


```

```python
plt.figure(figsize=(26,10))
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Comparacion entre Señal original y Señal filtrada')
plt.plot(data.X.to_numpy(), data.Y, color='gray')
plt.xlim(0,7500)
plt.xticks(range(0,7500,1000))
plt.plot(data.X.to_numpy() , savgol(data.Y,101,0) , label='101' , alpha=0.7,color='black' )
plt.savefig('Comparacion1.png')
```
