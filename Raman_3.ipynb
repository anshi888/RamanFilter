```python
import pandas as pd
filename = '/content/Punto 3 (1).csv'
data=pd.read_csv(filename,delimiter=';',decimal=",")

print(data)
```

```python
from numpy import *

x= data.X
y= data.Y
#pasar datos
x1= []
y1=[]
#print(x)
for c in range(len(x)):
    x1.append(float(x[c]))
    y1.append(float(y[c]))
#print(type(x1))
#print(x1)
```

```python
from numpy import *
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter as savgol

yf1 =  savgol(data.Y,31,7)

for u in range(len(y1)):
  data.Y[u] = yf1[u]
```

```python
print(data.Y)

data.to_csv('RAMAN3FILTER.csv',sep=',')
```

```python
from numpy import *
import matplotlib.pyplot as plt
# Y los graficamos
plt.figure(figsize=(26,10))
plt.title('Datos con ruido')
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim(100,3500)
plt.xticks(range(100,3500,150))
plt.plot(data.X.to_numpy(), data.Y)
plt.savefig('Se√±alOriginal3.png')
```

```python
from scipy.signal import savgol_filter as savgol
plt.figure(figsize=(32,10))
plt.title('se√±al filtrada ')
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim(0,3400)
plt.xticks(range(0,3400,150))

plt.plot(data.X.to_numpy() , savgol(data.Y,31,7) , label='31' , alpha=1 ,color='black')
plt.savefig('Se√±alfitrada3.png')
```
