```python
import pandas as pd
filename = '/content/Raman2 (1).csv'
data=pd.read_csv(filename,delimiter=',',decimal=",")
#pasar excel en formato numerico
```

Limpieza de datos

```python
from numpy import *

x= data.X
y= data.Y
#pasar datos
x1= []
y1=[]
#print(x)
for c in range(len(x)):
    x1.append(float(x[c]))
    y1.append(float(y[c]))
#print(type(x1))
#print(x1)
#print(y1)
```

```python
from numpy import *
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter as savgol

yf1 =  savgol(data.Y,81,2)

for u in range(len(y1)):
  data.Y[u] = yf1[u]
```

<h3>Grafico de señal original </h3>

```python

import matplotlib.pyplot as plt

plt.figure(figsize=(26,10))
plt.title('Datos con ruido',fontsize = 22)
plt.xlabel('X',fontsize = 22)
plt.ylabel('Y',fontsize = 22)
plt.ylim(0,1800)
plt.xlim(198,3400)
plt.plot(x1,y1,color='gray')
plt.savefig('señalorginal.png')
```

<h3>comparacion entre la señal orginal y filtrada</h3>

```python
from numpy import *
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter as savgol
```

```python
plt.figure(figsize=(30,13))
plt.xlabel('X',fontsize = 22)
plt.ylabel('Y',fontsize = 22)
plt.title('Comparacion entre Señal original y Señal filtrada',fontsize = 22)
plt.plot(x1,y1, color='gray')
plt.ylim(0,1800)
plt.xlim(198,3400)
plt.plot(x1 , savgol(data.Y,81,2) , label='81' , alpha=0.7,color='black' )
plt.savefig('comparacion2.png')
```

<h3>Señal Filtrada</h3>

```python
plt.figure(figsize=(30,13))
plt.xlabel('X',fontsize = 22)
plt.ylabel('Y',fontsize = 22)
plt.title('Señal filtrada',fontsize = 22)
plt.ylim(0,1800)
plt.xlim(198,3400)
plt.plot(x1 , savgol(data.Y,81,2) , label='81' , alpha=0.7,color='black' )
plt.savefig('señalfiltrada2.png')
```
