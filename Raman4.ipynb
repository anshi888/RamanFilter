```python
import pandas as pd
filename = '/content/Punto 4_1.csv'
data=pd.read_csv(filename,delimiter=',',decimal=",")

print(data)
```

```python
from numpy import *

x= data.X
y= data.Y
#pasar datos
x1= []
y1=[]
#print(x)
for c in range(len(x)):
    x1.append(float(x[c]))
    y1.append(float(y[c]))
#print(type(x1))
#print(x1)
```

```python
from numpy import *
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter as savgol

yf1 =  savgol(data.Y,71,9)

for u in range(len(y1)):
  data.Y[u] = yf1[u]
```

```python
print(data.Y)

data.to_csv('raman1filtrada4.csv',sep=',')
```

```python
from numpy import *
import matplotlib.pyplot as plt
# Y los graficamos
plt.figure(figsize=(26,10))
plt.title('Datos con ruido')
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim(0,8000)
plt.xticks(range(0,8000,1000))
plt.plot(data.X.to_numpy(), data.Y)
plt.savefig('Se√±alOriginal4.png')
```
